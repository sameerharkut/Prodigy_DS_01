```python
print(titanic_data.isnull().sum())
```

```python
titanic_data['Age'].fillna(titanic_data['Age'].median(), inplace=True)
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0], inplace=True)
```

```python
titanic_data.drop('Cabin', axis=1, inplace=True)
```

```python
print(titanic_data.isnull().sum())
```

```python
sns.countplot(x='Survived', data=titanic_data)
plt.title('Survival Distribution')
plt.show()

sns.countplot(x='Pclass', data=titanic_data)
plt.title('Passenger Class Distribution')
plt.show()

sns.countplot(x='Pclass', hue='Survived', data=titanic_data)
plt.title('Survival by Passenger Class')
plt.show()
```

```python
# Visualize the distribution of the 'Sex' variable
sns.countplot(x='Sex', data=titanic_data)
plt.title('Gender Distribution')
plt.show()

# Explore the relationship between 'Sex' and 'Survived'
sns.countplot(x='Sex', hue='Survived', data=titanic_data)
plt.title('Survival by Gender')
plt.show()
```

```python
# Visualize the distribution of the 'Age' variable
plt.hist(titanic_data['Age'], bins=20, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()

# Explore the relationship between 'Age' and 'Survived'
sns.boxplot(x='Survived', y='Age', data=titanic_data)
plt.title('Survival by Age')
plt.show()
```

```python
# Visualize the distribution of the 'SibSp' variable
sns.countplot(x='SibSp', data=titanic_data)
plt.title('Distribution of Siblings/Spouses Aboard')
plt.show()

# Explore the relationship between 'SibSp' and 'Survived'
sns.countplot(x='SibSp', hue='Survived', data=titanic_data)
plt.title('Survival by Siblings/Spouses Aboard')
plt.show()
```

```python
# Visualize the distribution of the 'Parch' variable
sns.countplot(x='Parch', data=titanic_data)
plt.title('Distribution of Parents/Children Aboard')
plt.show()

# Explore the relationship between 'Parch' and 'Survived'
sns.countplot(x='Parch', hue='Survived', data=titanic_data)
plt.title('Survival by Parents/Children Aboard')
plt.show()
```

```python
# Visualize the distribution of the 'Fare' variable
plt.hist(titanic_data['Fare'], bins=20, edgecolor='black')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Fare Distribution')
plt.show()

# Explore the relationship between 'Fare' and 'Survived'
sns.boxplot(x='Survived', y='Fare', data=titanic_data)
plt.title('Survival by Fare')
plt.show()
```

```python
# Explore the relationship between 'Embarked' and 'Survived'
sns.countplot(x='Embarked', hue='Survived', data=titanic_data)
plt.title('Survival by Embarked')
plt.show()
```
